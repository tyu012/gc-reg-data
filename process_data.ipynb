```python
import pandas as pd
```

```python
round1 = pd.read_csv("original/Course_Offering_0007_2022-11-6.csv")
```

```python
round1.head()
```

```python
scheduled_capacity = []
seats_r1 = []

for s in round1['SeatsAvailableScheduledCapacity']:
    cap = int(s.split('/')[1])
    fake_seats = int(s.split('/')[0])

    scheduled_capacity.append(cap)
    seats_r1.append(cap - (100 - fake_seats))

print(scheduled_capacity)
print(len(scheduled_capacity))

print(seats_r1)
print(len(seats_r1))
```

Note that for round 2 and 3 registration data, we will need to match the registration ID because the courses may not be presented in order.

```python
reg_data = round1[['REG_ID', 'SEC_CUR_STAT', 'SEC_NAME', 'SEC_TITLE', 'Days', 'Time', 'Faculty', 'CR_MIN']].copy()
reg_data['ScheduledCapacity'] = scheduled_capacity
reg_data['SeatsAvailableR1'] = seats_r1
reg_data.head()
```

```python
reg_data[reg_data['SEC_CUR_STAT'] != 'Cancelled'].to_markdown('output/reg_data.md')
```
